can be converted to ordinal values

Columns will be converted to ordinal values, or dummified<br>

```python
ordinal_columns = list()
non_ordinal_columns = list()

numeric_columns = list()
multivalued_columns = list()
```

```python
for column in data.columns:
    
    mapped = False
    
    if "LearningPlatformUsefulness" in column:
        data[column] = data[column].map({'Not Useful':1, 'Somewhat useful':2, 'Very useful':3})
        mapped = True
        
    if "Frequency" in column or "RemoteWork" in column:
        data[column] = data[column].map({'Rarely':1, 'Sometimes':2, 'Often':3, 'Most of the time':4})
        mapped = True
        
    if "TitleFit" in column:
        data[column] = data[column].map({'Poorly':1, 'Fine':2, 'Perfectly':3})
        mapped = True
        
    if "EmployerSize" in column:
        data[column] = data[column].map({
            'Fewer than 10 employees':1, 
            '10 to 19 employees':2, 
            '20 to 99 employees':3, 
            '100 to 499 employees':4, 
            '500 to 999 employees':5, 
            '1,000 to 4,999 employees':6, 
            '5,000 to 9,999 employees':7, 
            '10,000 or more employees':8, 
            "I don't know":np.nan, 
            'I prefer not to answer':np.nan
        })
        mapped = True
        
    if "JobSatisfaction" in column:
        data[column] = data[column].map({"Low":1, "Medium":2, "High":3})
        mapped = True
    
    if "Tenure" in column:
        data[column] = data[column].map({
            "I don't write code to analyze data":1,
            'Less than a year':2,
            '1 to 2 years':3,
            '3 to 5 years':4,
            '6 to 10 years':5,
            'More than 10 years':6            
        })
        mapped = True
    
    if "WorkDataVisualizations" in column:
        data[column] = data[column].map({
            'None':1,
            'Less than 10% of projects':2,
            '10-25% of projects':3,
            '26-50% of projects':4,
            '51-75% of projects':5,
            '76-99% of projects':6,
            '100% of projects':7
        })
        mapped = True
    
    if mapped:
        ordinal_columns.append(column)
    else:
        non_ordinal_columns.append(column)
```

```python
parameters_logr = [
    {
        'iimp__estimator'   : iimp_estimators,
        'logr__C'           : [0.1, 1.0],
        'logr__penalty'     : ['l1'],
        'logr__solver'      : ['liblinear'],
        'logr__class_weight' : [{0:0.2,1:1,1:1}, {0:0.4,1:1,1:1}, {0:0.6,1:1,1:1}, 'balanced']
    },
    {
        'iimp__estimator'   : iimp_estimators,
        'logr__C'           : [0.1, 1.0],
        'logr__penalty'     : ['l2'],
        'logr__solver'      : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
        'logr__class_weight' : [{0:0.2,1:1,1:1}, {0:0.4,1:1,1:1}, {0:0.6,1:1,1:1}, 'balanced']
    }
]
```
